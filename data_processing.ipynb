```python
import numpy as np
import pandas as pd
```

# Load original dataset. 

```python
data = pd.read_csv('SV_GQ_80.csv')
```

```python
data.head()
```

# Filtering out the missing values. 

```python
num_of_missings = (data == -1).sum()
L = num_of_missings[0].tolist()
average = sum(L) / len(L)
average
```

```python
x = (data == -1).sum()
d = pd.DataFrame({'Col':x.index, 'count':x.values})
y = d[d['count']<1]
l = y['Col'].to_list()
Final_df = data.filter(l)
Final_df
```

```python
# Final_df.to_csv('SV.csv', index=False)
```

```python
replaced_values = {
    0.0: np.array([1,0,0,0,0,0]), 
    1.0: np.array([0,1,0,0,0,0]), 
    2.0: np.array([0,0,1,0,0,0]), 
    3.0: np.array([0,0,0,1,0,0]),
    4.0: np.array([0,0,0,0,1,0]),
    5.0: np.array([0,0,0,0,0,1]),
    -1.0: np.array([0,0,0,0,0,0])
}
Final_df = Final_df.iloc[:, :-1].applymap(lambda idx: replaced_values[idx])
Final_df.head()
```

```python
np.array(Final_df.values.tolist(), dtype=np.float64).shape
```

```python
np.save(
    'sv_X.npy',
    np.array(Final_df.values.tolist())
)
np.save(
    'sv_Y.npy',
    np.array(data.iloc[:, -1].values.tolist())
)
```
